```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('score.csv')
```

```python
X = df['Hours']
y = df['Scores']

plt.scatter(X, y)
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# Reshape X to a 2D array using reshape(-1, 1)
# -1 means infer the number of rows, 1 means 1 column
model.fit(X.values.reshape(-1, 1), y)
```

```python
x_pred = model.predict([[9.25]])
print(x_pred)
```
