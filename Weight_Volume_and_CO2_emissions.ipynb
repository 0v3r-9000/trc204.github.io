```python
import pandas as pd

df = pd.read_csv("data.csv")

print(df.head())
print(df.info())
print(df.describe())
print(df.var())
```

```python
# Check for missing values
print(df.isnull().sum())

dropna = df.dropna()
print(dropna.info())

```

```python
X = df[["Weight","Volume"]]
y = df[["CO2"]]
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X,y)
```

```python
predictedDep = regr.predict([[1500, 837]])
```
